Реализовать паттерн Iterator (итератор) для множества.

```python
class Array:
    """
    Массив с операциями
    """
    def __init__(self, arr):
        """
        Конструктор
        """
        self._data = list(arr)

    def __len__(self):
        """
        Возвращает длину массива
        len -> int
        """
        return len(self._data)

    def __getitem__(self, item):
        """
        Возвращение элемента по индексу
        """
        return self._data[item]

    def __delitem__(self, key):
        """
        Удаление элемента по индексу
        """
        del self._data[key]

    def __add__(self, other):
        """
        Добавление элемента по индексу
        """
        self._data += [other]
        return self
    def returner(self):
        """
        Возвращение массива
        """
        return self._data

class SmallArray(Array):
    """Маленький массив"""
    def __init__(self, arr):
        super().__init__(arr)

class BigArray(Array):
    """Большой массив"""
    def __init__(self, arr):
        super().__init__(arr)

    def __getitem__(self, item):
        "Возвращает элемент в двоичном виде "

        return bin(self._data[item])

class Set:
    maxs_len_small_array = 100 #максимальная длина маленького массива

    def __init__(self, arr, cls=SmallArray):
        """
        Конструктор множества
        arr - массив
        cls - экземпляр маленького массива
        """
        self.cls = cls
        if len(arr) > Set.maxs_len_small_array:
            self.cls = BigArray
        else:
            self._data = self.cls(arr)

    def add(self, value):
        """
     Добавляет элемент в множество

        """
        if self.cls != BigArray and (len(self._data) >=
                                     Set.maxs_len_small_array):
            self.cls = BigArray
            self._data = self.cls(self._data.returner())
        self._data += value

    def delete(self, index):
        """
    Удаляет элемент из множества по индексу

        """
        if self.cls == BigArray and (len(
                self._data) == Set.maxs_len_small_array + 1):
            self.cls = SmallArray
            self._data = self.cls(self._data)
        del self._data[index]

    def get(self, index):
        """
     Возвращает элемент множества по индексу

        """
        return self._data[index]

class IterableSet(Set):
    """
    Множество с итератором
    """
    def __iter__(self):
        """
        Функция итерирования
        """
        self.__n = -1
        return self

    def __next__(self):
        """
        Возвращение элемента при итерации
        """
        if self.__n == len(self._data) - 1:
            raise StopIteration
        self.__n += 1
        return self.get(self.__n)
```

```python
print("Создадим маленький массив")
first = IterableSet([3, 2, 1, 6, 9])
for i in first:
    print(i)
print("Создадим маленький массив расширим его")
second = IterableSet(range(100))
second.add(22)
for i, j in enumerate(second):
    if i < 13:
        print(j)
```
